```python
import plotly.graph_objects as go
import pandas as pd

# df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
df = pd.read_csv('Final.csv')

print(df)
```

```python
sub = df[["artist", "title", "total_streams"]]
print(sub)
```

```python

f = {'total_streams': 'sum', 'artist': 'first' }
temp = sub.groupby(['title']).agg(f)
temp1 = temp.loc[temp['artist'] == 'BTS']
sorted_df = temp1.sort_values(by=['total_streams'], ascending=False)
df_first_10 = sorted_df.head(10)
print(df_first_10)
```

```python
import seaborn as sns
import matplotlib as math
import matplotlib.pyplot as plt


df_first_10_all_col = df_first_10.reset_index()

plt.style.use("dark_background")

gfg = sns.barplot(x = 'total_streams',y = 'title', color = "#1ed760" ,data = df_first_10_all_col,errwidth=0)

plt.title("Most Popular Songs", fontsize=15)
gfg.set_xlabel('Stream Count',fontsize=12, color='w')
gfg.set_ylabel('Title',fontsize=12, color='w')
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```
